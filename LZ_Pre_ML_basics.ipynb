## What is the score?
- Accuracy no longer make sense (# correct / # total)
- Doen't make sense for regression
  - If true house price is 1 million dollars, and we predict 1 miliion and 1 dollar. - are we wrong?
  - Intuitively we know it's pretty good.
  - If we predict 2 million then it's bad
- Regression uses MSE(Mean Squared Error). But it's not returned by score()
  - Y with the "hat" on top is the prediction

$$
MSE = 1/N \sum_{i=1}^{N}(\hat{Y_i} - Y_i)^2
$$

### Why not MSE?
- The house prices will range hundred of thousands to millions
- The grades will range between 0..100
- An error of $100^2$ is not bad for house prices, but very bad for grades
- The relative scale matters

### The $\ $ $R^2$
- Is literally corelation coefficent
- SSE: Sum of squared errors
  - For each predition, we take the difference between it and the corresponding target, square difference and add all squared differnces together
- SST: Sum of the squared total
  - The sum of squared differences between each target and the sample mean of the targets
  - It's proportional to the variance of the targets
  - If we divide SST by N, it's $1 - \frac{MSE}{variance}$

- Perfect model: $\ SSE = 0, \ R^2 = 1 $
- Dumbest model: Predict average every time, $\ SSE= SST, \ R^2 = 0$
- $\ R^2 < 0$, if predictions are so widely incorrect, $\ SSE> SST $, it's even worse then dumbest model

$$
$$

$$
R^2 = 1\  - \ \frac{SSE}{SST}
$$
$$
SSE = \frac{1}{N} \sum_{i=1}^{N}(y_i - \hat{y_i} )^2
$$
$$
SST = \sum_{i=1}^{N}(y_i - \bar{y} )^2, \bar{y}\ means\ mean\ of\ all\ y^{'}
$$
