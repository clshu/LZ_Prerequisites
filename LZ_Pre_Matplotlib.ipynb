# Exercise

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt

# Make 1D arrays for x and y axes and spread them around -1 and 1
x1 = (np.random.random(2000)-0.5)*2
x2 = (np.random.random(2000)-0.5)*2

# Make an array for color picking
Y = np.zeros(2000)

for e in range(len(Y)):
  if (x1[e] <= 0 and x2[e] > 0) or (x1[e] > 0 and x2[e] <= 0):
    Y[e] = 1

# Plot it all
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Matplotlib Exercise')
plt.scatter(x1,x2,c=Y);
```

## Objective Functions
- ### When training machine learning models, we start with **objective**.
- ### Objective != "I want to build self-driving cars"
- ### Objective = a number we want to minimize or maximize, when you reach your optimal number, you can say your training is complete.


- e.g. "Mean Squared Error"
- If the line fit MSE is smaller, otherwise, larger.

$$
MSE = 1/n \sum_{i=1}^{n}(\hat{Y_i} - Y_i)^2
$$

##  Classification Objective

- Error Rate = (# I got wrong) / (# total pridiction)

## Plotting Objective Per Iteration
- "Training" a machine learning model often involes an **iterative porcess** whereby the objective is improved on each iteration
- We can check whether training was successful by plotting the objective

### Examples of iterative algorithms:
- Gradient Descent
- AKA Backpropagation (neural networks)
- Expectation-Maximization (probabilistic latent variable models)
- Coordinate descent

### Examples of models:
- Logistic regression
- Deep neural networks
- Hidden Markov models
- Matric factorization
- Reenforced Learning (it builds an agent that acts in real time)
  - Play chess, go and vedio games
  - Reinforcement learning agents are trained by playing games over and over again, and each of these games is called an episode.
  - So this is where we get back to this idea of plotting in order to make sure that our agent is learning the way we intend it to, we have to plot the total reward per episode.
  - And again, you should recognize that this is just a basic line chart.
  - So again, very complex ideas, but all we need are the basic plots you've already learned about.
  - Finance -> Time series -> Line Chart
  - Stock Return -> Histgram -> A fat tail distribution
  - Computer vision -> self-driving cars
  - Deep Learning
  - Gans generative adversarial networks

- Most of the time, although this may be surprising to you, all we need are basic plots like the line chart, the scatter plot, the histogram and plotting images.
